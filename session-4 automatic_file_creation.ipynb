```python
import matplotlib.pyplot as plt

n = len(cat)  # total number of categorical columns

plt.figure(figsize=(24, 10))  # figure का size set करो

for i in range(1, n):  # cat[1:] से शुरू किया है
    plt.subplot(3, 3, i)  # 3x3 grid में subplot
    col = cat[i]  # column name
    
    keys = visa_df[col].value_counts().keys()
    values = visa_df[col].value_counts().values

    plt.bar(keys, values, color='green')  # bar chart
    plt.title(f'Bar Chart of {col}')  # chart title
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)  # x labels को घुमाओ

plt.tight_layout()  # spacing adjust करो
plt.show()
```
